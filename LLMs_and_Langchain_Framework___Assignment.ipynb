## Question 1: What are Large Language Models (LLMs) and how do they function?  

**Answer:**  

Large Language Models (LLMs) are advanced AI models trained on massive text datasets to understand and generate human-like language.

### How they function:
- Based on **Transformer architecture**
- Use **self-attention mechanism** to understand context  
- Trained on large datasets using **unsupervised learning**  
- Predict the **next word/token** in a sequence  
- Examples include GPT, BERT, etc.

**Question 8:** Create a LamaIndex implementation that indexes a local text file and
retrieves answers from it.

````python
!pip install llama-index
!pip install llama-index-embeddings-huggingface
!pip install llama-index-llms-huggingface
````

````python
import os

os.makedirs("data", exist_ok=True)

with open("data/sample.txt", "w") as f:
    f.write("Artificial Intelligence helps machines learn from data and make decisions.")
````

````python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.huggingface import HuggingFaceLLM

# Local embedding
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en")

# FIXED LLM (causal model)
Settings.llm = HuggingFaceLLM(model_name="gpt2")

# Load documents
documents = SimpleDirectoryReader("data").load_data()

# Create index
index = VectorStoreIndex.from_documents(documents)

# Query
query_engine = index.as_query_engine()
response = query_engine.query("What is Artificial Intelligence?")

print(response)
````

**Question 9:** Demonstrate combining Langchain with LamaIndex to create a simple
document-based Q&A chatbot.


**Answer:**

In this implementation, we combine LangChain and LlamaIndex to build a simple document-based Question Answering (Q&A) chatbot.

LlamaIndex is used to load and index the documents, converting them into vector embeddings for efficient retrieval. LangChain is then used to manage the interaction with the language model and generate responses based on retrieved context.

Steps involved:
1. Load local documents using SimpleDirectoryReader.
2. Create a vector index using LlamaIndex.
3. Use a local HuggingFace embedding model to avoid API key dependency.
4. Configure a lightweight HuggingFace LLM.
5. Convert the index into a query engine.
6. Pass user queries to retrieve context-aware answers.

This combination allows us to build an intelligent chatbot that answers questions based only on the provided documents.

````python
# Install
!pip install llama-index langchain

# Imports
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.llms.huggingface import HuggingFaceLLM

# Set LLM (IMPORTANT FIX)
Settings.llm = HuggingFaceLLM(
    model_name="gpt2",
    tokenizer_name="gpt2",
    max_new_tokens=100
)

# Load documents
documents = SimpleDirectoryReader("/content/data").load_data()

# Create index
index = VectorStoreIndex.from_documents(documents)

# Query engine
query_engine = index.as_query_engine()

# Ask question
response = query_engine.query("What is this document about?")
print(response)
````

**Question 10:** A legal firm wants to use AI to summarize large volumes of legal
documents and retrieve relevant information quickly. Propose a solution using
Langchain and LamaIndex, and explain how it would work in practice.



## Overview
A legal firm needs an AI system to:
- Summarize large volumes of legal documents
- Quickly retrieve relevant information

This can be achieved by combining:
- **LlamaIndex** → for document indexing and retrieval
- **LangChain** → for processing, chaining, and interaction with LLMs

---

## Proposed Solution Architecture

### 1. Data Ingestion
- Load legal documents (PDFs, Word files, text files)
- Use document loaders to read and preprocess data

```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader("legal_docs").load_data()

# 2. Indexing (LlamaIndex)

Documents are split into smaller chunks\
Each chunk is converted into embeddings\
Stored in a vector index for fast search

```
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()
```

# 3. Retrieval (LlamaIndex)
User query is matched with document embeddings\
Most relevant chunks are retrieved

```
response = query_engine.query("Explain termination clauses")
print(response)
```

# 4. Processing & Chaining (LangChain)

LangChain handles prompt creation and LLM interaction\
Chains multiple steps (retrieval → summarization)
```
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI

prompt = PromptTemplate(
    input_variables=["context"],
    template="Summarize the following legal text:\n{context}"
)

llm = OpenAI()
chain = LLMChain(llm=llm, prompt=prompt)

summary = chain.run(context=str(response))
print(summary)

# 5. Summarization

* Retrieved legal content is summarized using LLM\
* Extracts:\
  *  Key clauses
  * Important legal points
  * Concise explanations
